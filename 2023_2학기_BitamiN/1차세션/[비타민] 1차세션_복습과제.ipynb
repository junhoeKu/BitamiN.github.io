```python
# 1. spotify.csv를 spotify라는 데이터 프레임으로 불러온 후 상위 5행을 출력하세요

spotify = pd.read_csv('/content/spotify.csv')
spotify.head()
```

```python
# 2. spotify 데이터의 분포도를 확인하세요

spotify.describe()
```

```python
# 3. spotify 데이터의 결측치를 확인하세요

spotify.isna().sum()

# spotify.isna() 보다 .sum()을 한 번 더 취해줌으로써 결측치를 효율적으로 확인하는 것을 더 추천.
```

```python
# 4. spotify에서 'Artist'열의 데이터 분포를 확인하세요.

spotify['Artist'].value_counts()
```

```python
# 5. spotify에서 Artist가 Post Malone인 데이터만 posty라는 새로운 데이터프레임으로 만든 후 출력하세요

posty = spotify[spotify['Artist']=='Post Malone']
# 또는
# posty = spotify.loc[spotify['Artist']=='Post Malone'] or posty = spotify.loc[spotify['Artist']=='Post Malone', :]
posty
```

```python
# 6. posty 데이터프레임에서 Rank의 최소값을 구하세요

posty['Rank'].min()
```

```python
# 7. posty 데이터프레임에서 tempo의 중앙값을 구하세요

posty['tempo'].median()
```
