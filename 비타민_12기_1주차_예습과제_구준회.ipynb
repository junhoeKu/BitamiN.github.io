**1. Matplotlib의 pyplot 모듈을 사용하여 시각화해봅니다.**

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

#### 1) Figure과 facecolor을 활용하여 plot을 생성하기.


```python
# figsize 크기가 가로 4, 세로 4이며 배경색이 노란색인 plot을 반환하세요.
plt.figure(figsize = (4,4), facecolor = 'yellow')
plt.plot([1,2,3], [2,4,6])
plt.show()
```

#### 2) Axes 객체를 이용하여 선 그래프 그리기.


```python
# plt.subplots를 이용하여 Figure와 Axes 객체를 함께 가져옵니다. Figsize는 (4,4)로 설정
fig, ax = plt.subplots(figsize=(4, 4))

# Axes 객체의 plot()을 이용하여 선 그래프를 그립니다.
ax.plot([1,2,3],[2,4,6])


# Axes 객체의 set_title()을 이용하여 선 그래프의 타이틀을 자유롭게 설정하세요.
ax.set_title('Line Plot')

plt.show()
```

#### 3) 여러 개의 plot을 subplot들을 생성하기.

```python
# 두 개의 subplot, 즉 두 개의 Axes를 생성하되 열 방향으로 두 개를 배치함.
# 또한 전체 크기가 가로 6, 세로 3인 Figure 생성.
fig, ax = plt.subplots(2, 1, figsize=(6, 3))
plt.show()
```

#### 4) pyplot의 plot() 함수를 이용하여 간단한 직선 그래프 그리기.

```python
# x_value에 [1,2,3,4], y_value에 [2,4,5,8]을 가지는 파이썬 리스트를 생성하고
# plot()에 인자로 입려하여 간단한 직선 그래프를 만들어 보세요.
# color 인자를 설정하여 선의 색깔을 녹색으로 변경합니다.

x_value = [1, 2, 3, 4]
y_value = [2, 4, 5, 8]

plt.plot(x_value, y_value, color='green')

plt.show()
```

#### 5) 축의 눈금값 회전, 범례 설정하기.

```python
# x 좌표 값을 0~99까지 100개로 설정.
x_value=np.arange(0,100)

# y 좌표 값은 개별 x좌표 값을 2배로 해서 100개로 설정.
y_value = x_value * 2

plt.plot(x_value, y_value, color='green')

# x축 눈금값을 90도로 회전
plt.xticks(rotation=90)
plt.show()
```

```python
x_value_01 = np.arange(1,10)
y_value_01 = 2*x_value_01

# 마커를 포함한 빨간색 대시 선 그래프를 그리고, label은 'red line'으로 설정
plt.plot(x_value_01, y_value_01, linestyle='--', color='red', label='red line')

# x값에 따른 y값을 나타내는 초록색 막대 그래프를 그리고 label은 'bar plot'으로 설정
plt.bar(x_value_01, y_value_01, color='green', label='bar plot')

plt.xlabel('x axis')
plt.ylabel('y axis')

# 개별 plot들의 label로 설정된 문자열을 범례로 표시
plt.legend()
plt.show()
```

#### 6) 여러 개의 subplot을 이용해 개별 그래프를 subplot별로 시각화하기

```python
# 다음 plot과 같은 형태로 출력하세요. figsize는 (6,3)
fig, ax = plt.subplots(1, 2, figsize=(6, 3))
```

```python
x_value_01 = np.arange(1,10)
x_value_02 = np.arange(1,20)
y_value_01 = 2 * x_value_01
y_value_02 = 2 * x_value_02

fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(6,3))

# 개별 Axes 객체를 이용하여 선 그래프와 막대그래프를 그립니다.
# ?를 채워넣으세요
ax[0].plot(x_value_01, y_value_01, color='red', marker='o', linestyle='dashed', label='red line')
ax[1].bar(x_value_02, y_value_02, color='green', label='bar plot')

# 개별 Axes 객체 각각에 X축명을 설정.
ax[0].set_xlabel('ax[0] x axis')
ax[1].set_xlabel('ax[1] x axis')

# 개별 Axes 객체 각각에 범례 설정
ax[0].legend()
ax[1].legend()

plt.show()
```

**2. Seaborn을 사용하여 시각화해봅니다.**

#### 0) 데이터 불러오기(실행해 주세요.)

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings("ignore")

titanic_df=pd.read_csv("../2024_0학기_비타민/titanic.csv")
titanic_df = titanic_df.dropna(subset=["Embarked"])
titanic_df.head(5)
```

#### 1) 여러가지 컬럼들로 시각화하기
subplot으로 2*3 개의 subplots만들기 figsize=(14, 8)  

(0, 0) age의 히스토그램을 시각화, survived로 색깔을 달리하기, 연속 확률분포 곡선도 표현하기   
(0, 1) pclass에 따른 age를 수평 barplot으로 만들기, 99% 신뢰구간을 표시하기   
(0, 2) pclass에 따른 survived의 합을 barplot으로 만들기, sex로 색깔을 달리하기(신뢰구간 표시 x)   
(1, 0) embarked에 따른 fare의 boxplot 만들기, Pclass로 색깔을 달리하기  
(1, 1) survived에 따른 age의 violin plot 만들기, sex로 색깔을 달리하기  
(1, 2) age에 따른 fare의 scatter plot 만들기, Survived로 색깔을 달리하고 Embarked로 모양을 달리해 표현하기

```python
fig, axs = plt.subplots(nrows=2, ncols=3, figsize=(14, 8))

sns.histplot(x='Age', data=titanic_df, ax=axs[0, 0], hue='Survived', kde = True)
sns.barplot(x='Age', y='Pclass', data=titanic_df, ax=axs[0, 1], ci=99, orient='h')
sns.barplot(x='Pclass', y='Survived', data=titanic_df, ax=axs[0, 2], hue='Sex', estimator = 'sum', ci = False)
sns.boxplot(x='Embarked', y='Fare', data=titanic_df, ax=axs[1, 0], hue='Pclass')
sns.violinplot(x='Survived', y='Age', data=titanic_df, ax=axs[1, 1], hue='Sex')
sns.scatterplot(x='Age', y='Fare', data=titanic_df, ax=axs[1, 2], hue='Survived', style='Embarked')
plt.show()
```

#### 2) 상관 히트맵 만들기

1. 숫자형 컬럼만 살리기

2. 상관계수의 숫자를 소수점 두번째 자리까지 표시하기

```python
# 숫자형 데이터만을 가진 데이터프레임 생성
titanic_df_numeric = titanic_df.select_dtypes(include=['float64', 'int64'])

# 상관 히트맵 생성
sns.heatmap(titanic_df_numeric.corr().round(2), annot=True)

plt.show()
```
